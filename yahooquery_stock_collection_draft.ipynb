```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import yahooquery as yq
from yahooquery import Ticker
from datetime import datetime, timedelta

symbols = ['aapl', 'amzn', 'nflx', 'goog']
tickers = Ticker(symbols)
```

```python
from datetime import datetime, timedelta
# 기간
interval='1m'
end = datetime.today()
start = end-timedelta(days=7)
a = tickers.history(start=start, end=end, interval=interval)
a

# end = datetime.today()
# for i in range(1, 5):
#     start = end-timedelta(days=7)
#     a = tickers.history(start=start, end=end, interval=interval)
#     end = start

# a
```

```python
# 멀티인덱스 수정
company_datas = dict()
for company in symbols:
    tmp_df = a.loc[company]
    company_datas[company] = tmp_df
company_datas
```

```python
df = pd.DataFrame()
for company in symbols:
    company_datas[company]["date"] = company_datas[company].index
    company_datas[company]["symbol"] = company
    df = pd.concat([df, company_datas[company]], join='outer')
df.reset_index(drop=True, inplace=True)
df
```
