```python
import pandas as pd
```

```python
kospi = pd.read_csv("./코스피기본정보.csv", encoding='euc-kr')
nasdaq = pd.read_csv("./나스닥기본정보.csv", encoding='euc-kr')
additional = pd.read_csv("./asset_profile.csv", encoding='utf-8')
```

```python
all = pd.concat([kospi, nasdaq], axis=0)
all
```

```python
all = all.drop(columns='Unnamed: 0')
all
```

```python
all = all.reset_index()
all = all.drop(columns='index')
all
```

```python
# 코스피는 뒤에 '.ks'
kospi_condition = (all['market']=='kospi')
kospi_condition
all.loc[(kospi_condition), "code"] = all['code']+".KS"
all
```

```python
all.rename(columns = {'phone' : '전화번호','website' : '사이트주소'}, inplace = True)
```

```python
additional.rename(columns = {'symbol_id' : 'code'}, inplace = True)
# additional.columns
additional = additional.drop(columns=['Unnamed: 0', 'country', 'industry', 'sector'])
additional
```

```python
r = pd.merge(all,additional, how='outer')
r
```

```python
r = r.drop(columns=['전화번호', '사이트주소'])
r
```
