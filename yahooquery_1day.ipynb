```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import yahooquery as yq
from yahooquery import Ticker
from datetime import datetime, timedelta

# s1 = pd.read_csv("./all_stocks_basic_data.csv", encoding='euc-kr')
# s1 = s1.iloc[:, 1]
# s1 = s1.values
# symbols = s1.tolist()
# s1
symbols = ['aapl', 'amzn', 'nflx', 'goog']
tickers = Ticker(symbols)
symbols
```

```python
# 1일단위 데이터 추출
interval='1d'
start = '1997-01-01'
end = datetime.today()
df = pd.DataFrame()
a = tickers.history(start=start, end=end, interval=interval)
# 멀티인덱스 수정
company_datas = dict()
for company in symbols:
    tmp_df = a.loc[company]
    company_datas[company] = tmp_df
    company_datas[company]["date"] = company_datas[company].index
    company_datas[company]["symbol"] = company
    df = pd.concat([df, company_datas[company]], axis=0)
df
```

```python
df = df.drop(df.columns[[-2]], axis=1)
# df.to_csv("C:/Users/SSAFY/Downloads/drafts2.csv", sep=',', encoding='utf-8')
df
```
