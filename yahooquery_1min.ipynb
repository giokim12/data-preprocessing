```python
s1 = pd.read_csv("./valid_company_list.csv", encoding='euc-kr')
s1 = s1.iloc[:, 1]
s1 = s1.values
symbols = s1.tolist()
print(len(symbols))
symbols.remove('030790.KS')
symbols.remove('007610.KS')
symbols.remove('012600.KS')
symbols.remove('003620.KS')
symbols.remove('015540.KS')
symbols.remove('109070.KS')
symbols.remove('019490.KS')
# symbols.remove('003620.KS')

print(len(symbols)) 
tickers = Ticker(symbols,asynchronous=True)
# symbols
```

```python
# 1분단위 데이터 추출
interval='1m'
end = datetime.today()
df = pd.DataFrame()
for i in range(1, 5):
    start = end-timedelta(days=7)
    a = tickers.history(start=start, end=end, interval=interval)
    print(f'1분단위 7일 데이터 호출 성공, {i}번째')
    # 멀티인덱스 수정
    company_datas = dict()
    for company in symbols:
        tmp_df = a.loc[company]
        company_datas[company] = tmp_df
        company_datas[company]["date"] = company_datas[company].index
        company_datas[company]["symbol"] = company
        df = pd.concat([df, company_datas[company]], axis=0)
        print('데이터 합치기 성공', company)
    end = start
    
df
```
