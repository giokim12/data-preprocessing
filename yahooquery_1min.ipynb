```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import yahooquery as yq
from yahooquery import Ticker
from datetime import datetime, timedelta

s1 = pd.read_csv("./valid_company_list.csv", encoding='euc-kr')
s1 = s1.iloc[:, 1]
s1 = s1.values
symbols = s1.tolist()
# s1
# symbols = ['aapl', 'amzn', 'nflx', 'goog']
tickers = Ticker(symbols)
symbols
```

```python
# 1분단위 데이터 추출
interval='1m'
end = datetime.today()
df = pd.DataFrame()
for i in range(1, 5):
    start = end-timedelta(days=7)
    a = tickers.history(start=start, end=end, interval=interval)
    print(f'7일동안 1분단위 데이터 추출성공 {i}주차' )
    # 멀티인덱스 수정
    company_datas = dict()
    for company in symbols:
        print(company)
        tmp_df = a.loc[company]
        company_datas[company] = tmp_df
        company_datas[company]["date"] = company_datas[company].index
        company_datas[company]["symbol"] = company
        df = pd.concat([df, company_datas[company]], axis=0)
        print(f'{i}주차 {company}회사' )
    end = start
    
df
```
