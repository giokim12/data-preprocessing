```python
import pandas as pd
pd.set_option('display.max_row', 60)
pd.set_option('display.max_columns', 10)
```

```python
raw_data_k = pd.read_csv("./korea_interest_rate.csv", encoding='utf-8')
```

```python
raw_data_k = raw_data_k.drop(raw_data_k.columns[[0,1,2,3]], axis=1)
raw_data_k
```

```python
df = raw_data_k.transpose()
df = df.reset_index()
df
```

```python
df = df.rename(columns={'index':'date', 0:'korea'})
df
```

```python
df.dtypes
df['date'] = pd.to_datetime(df['date'])
df
```

```python
raw_data_u = pd.read_csv("./usa_interest_rate_raw.csv", encoding='utf-8')
raw_data_u
```

```python
df_u = raw_data_u.rename(columns={'observation_date':'date', 'FEDFUNDS':'usa'})
df_u['date'] = pd.to_datetime(df_u['date'])
df_u = df_u[df_u['date'] >= '1997-01-01']
df_u
```

```python
df_OUTER_JOIN = pd.merge(df_u, df, left_on='date', right_on='date', how='left')
df_OUTER_JOIN
```

```python
# pd.set_option('display.max_row', 500)
# pd.set_option('display.max_columns', 100)
```

```python
df_OUTER_JOIN
```

```python
MERGED = pd.merge(df, df_u, how = 'outer')
MERGED
```

```python
C = pd.concat([df_u, df])
C = C.sort_values(by=['date'], axis=0, ascending=False, inplace=False)
C
```

```python
d = C.drop_duplicates(subset='date')
d
```

```python
e = pd.merge(d, df, left_on='date', right_on='date', how='left')
e
```

```python
f = pd.merge(e, df_u, left_on='date', right_on='date', how='left')
f
```

```python
g = f.drop(['korea_x', 'usa_x'], axis=1)
g
```

```python
g = g.rename(columns={'usa_y':'usa', 'korea_y':'korea'})
g
```
