```python
import pandas as pd
import numpy as np
pd.set_option('display.max_row', 60)
pd.set_option('display.max_columns', 10)
```

```python
raw_data_k = pd.read_csv("./korea_interest_rate.csv", encoding='utf-8')
```

```python
raw_data_k = raw_data_k.drop(raw_data_k.columns[[0,1,2,3]], axis=1)
raw_data_k
```

```python
df = raw_data_k.transpose()
df = df.reset_index()
df
```

```python
df = df.rename(columns={'index':'date', 0:'korea'})
df
```

```python
df.dtypes
df['date'] = pd.to_datetime(df['date'])
df
```

```python
raw_data_u = pd.read_csv("./usa_interest_rate_raw.csv", encoding='utf-8')
raw_data_u
```

```python
df_u = raw_data_u.rename(columns={'observation_date':'date', 'FEDFUNDS':'usa'})
df_u['date'] = pd.to_datetime(df_u['date'])
df_u = df_u[df_u['date'] >= '1997-01-01']
df_u
```

```python
exc_raw = pd.read_csv("./exchange_rate.csv", encoding='utf-8')
exc_raw
```

```python
exc = exc_raw.rename(columns={'Close':'exchange_rate', 'Date':'date'})
exc = exc.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume', 'exchange'], axis=1)
exc = exc.astype({ 'date' : 'datetime64[ns]' })
exc
```

```python
exc.dtypes
```

```python
exc_and_korea = pd.merge(df, exc, left_on='date', right_on='date', how='left')
exc_and_korea
```

```python
df_OUTER_JOIN = pd.merge(exc_and_korea, df_u, left_on='date', right_on='date', how='left')
df_OUTER_JOIN
```

```python
# pd.set_option('display.max_row', 500)
# pd.set_option('display.max_columns', 100)
```

```python
df_OUTER_JOIN
```

```python
df_OUTER_JOIN['usa'] = pd.to_numeric(df_OUTER_JOIN['usa'], errors='coerce').fillna(0)
rate = 0
for i in range(len(df_OUTER_JOIN['usa'])):
    if df_OUTER_JOIN['usa'][i] != 0:
        rate = df_OUTER_JOIN['usa'][i]
        print(rate, '값')
    elif df_OUTER_JOIN['usa'][i] == 0:
        df_OUTER_JOIN['usa'][i] = rate
        print('값바꿈')

df_OUTER_JOIN
```

```python
df_OUTER_JOIN['exchange_rate'] = pd.to_numeric(df_OUTER_JOIN['exchange_rate'], errors='coerce').fillna(0)
df_OUTER_JOIN
```

```python
df_OUTER_JOIN.to_csv('./exchange_interest.csv', sep=',', encoding='utf-8')
```
