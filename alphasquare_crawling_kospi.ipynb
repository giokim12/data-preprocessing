```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
from bs4 import BeautifulSoup
import requests
import ast
from ast import literal_eval
import json
```

```python
# 코스피정보
kospi = pd.read_csv("./kospi.csv", encoding='euc-kr')
kospi = kospi.drop(['종가', '대비', '등락률', '시장구분'], axis=1)
kospi
```

```python
codes = kospi.iloc[:, 0]
codes= codes.values
codes = codes.tolist()
df = pd.DataFrame(columns = ['code','logo','ko_name','description', 'sector', 'industry', 'country_code', 'phone', 'website', 'market'])
df
```

```python
res = requests.get(f'https://api.alphasquare.co.kr/data/v2/stock/details?code=AAPL')
response = res.text
print(response) 
```

```python
ind = range(1)
for i in range(len(codes)):
    res = requests.get(f'https://api.alphasquare.co.kr/data/v2/stock/details?code={codes[i]}')
    response = res.text
    print(response)
    res_dict = dict(json.loads(response))
    print(type(res_dict[codes[i]]['description']))
    if type(res_dict[codes[i]]['description']) == str:
        res_dict[codes[i]]['description'] = res_dict[codes[i]]['description'].replace("\r\n\r\n"," ")
    if res.status_code == 200:
        print(f'{i} of 215')
        new_data = {'code' :res_dict[codes[i]]['code'],'logo': res_dict[codes[i]]['logo'], 'ko_name': res_dict[codes[i]]['ko_name'],'description': res_dict[codes[i]]['description'],'sector': res_dict[codes[i]]['sector'],'industry':res_dict[codes[i]]['industry'],'country_code':res_dict[codes[i]]['country_code'],'phone':res_dict[codes[i]]['phone'],'website':res_dict[codes[i]]['website'],'market':res_dict[codes[i]]['market']}
        # df = df.append(new_data, ignore_index=True)
        new_data = pd.DataFrame(new_data, index =ind)
        df = pd.concat([df, new_data], axis = 0)
        # print(df)
df
```
