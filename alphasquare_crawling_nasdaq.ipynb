```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime
from bs4 import BeautifulSoup
import requests
import ast
from ast import literal_eval
import json
```

```python
# 나스닥에서 mega랑 large만 필터링에서 다 다운로드받은거(214개 기업)
nasdaq = pd.read_csv("./nasdaq.csv")
nasdaq = nasdaq[['Symbol', 'Name', 'IPO Year', 'Sector', 'Market Cap']]
nasdaq = nasdaq.rename(columns={"Symbol":"종목코드", 'Name':'종목명', 'IPO Year':'상장일', 'Sector':'업종명', 'Market Cap':'시가총액'})
nasdaq['시장구분'] ='NASDAQ'
nasdaq['한글 종목명'] = nasdaq['종목명']
```

```python
codes = nasdaq.iloc[:, 0]
codes= codes.values
codes = codes.tolist()
df = pd.DataFrame(columns = ['code','logo','ko_name','description', 'sector', 'industry', 'country_code', 'phone', 'website'])
df
```

```python
ind = range(1)
pd.set_option('display.max_row', 300)
pd.set_option('display.max_columns', 10)
for i in range(len(codes)):
    res = requests.get(f'https://api.alphasquare.co.kr/data/v2/stock/details?code={codes[i]}')
    response = res.text
    res_dict = dict(json.loads(response))
    if type(res_dict[codes[i]]['description']) == str:
        res_dict[codes[i]]['description'] = res_dict[codes[i]]['description'].replace("\r\n\r\n"," ")
    if res.status_code == 200:
        new_data = {'code' :res_dict[codes[i]]['code'],'logo': res_dict[codes[i]]['logo'], 'ko_name': res_dict[codes[i]]['ko_name'],'description': res_dict[codes[i]]['description'],'sector': res_dict[codes[i]]['sector'],'industry':res_dict[codes[i]]['industry'],'country_code':res_dict[codes[i]]['country_code'],'phone':res_dict[codes[i]]['phone'],'website':res_dict[codes[i]]['website']}
        new_data = pd.DataFrame(new_data, index =ind)
        df = pd.concat([df, new_data], axis = 0)
        # print(df)
df
```
